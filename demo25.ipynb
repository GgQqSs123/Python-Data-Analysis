```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 导入本节示例数据，注意此文件并非本地数据文件，而是从Seaborn库里导入的
# Seaborn库里的所有数据可以在这里查看：https://github.com/mwaskom/seaborn-data
# 如果遇到导入问题，可以从以上GitHub仓库下载到本地，然后读取文件
# 或者是参考https://blog.csdn.net/weixin_51755670/article/details/127979895
df1 = sns.load_dataset("penguins").query('species == "Adelie"')
s1 = df1['bill_length_mm']
s1
```

## 对Series进行绘图

```python
# 把s1绘制为直方图
# sns.histplot

sns.histplot(s1)
# 调用matploblib plt.show()
plt.show()
```

```python
# 把s1绘制为密度图
# sns.kdeplot
sns.kdeplot(s1)
plt.show()
```

```python
# 把s1绘制为箱型图
# sns.boxplot
sns.boxplot(s1)
plt.show()
```

```python
# 把s1绘制为小提琴图
# sns.violinplot
sns.violinplot(s1)
plt.show()
```

## 给图表添加标题和轴标签

注意：Matplotlib当前默认的字体并不支持中文展示，所以我们需要把字体换成其它支持中文的字体。
由于每台电脑上有的字体并不系统，你可以先查询当前系统所有字体，然后把字体替换成其中某个支持中文的字体。

```python
# 查询当前系统所有字体
import matplotlib
from matplotlib.font_manager import FontManager

mpl_fonts = set(f.name for f in FontManager().ttflist)
for f in sorted(mpl_fonts):
    print(f)
    # 把df1的"bill_length_mm"列绘制为直方图
# 把图表标题设置为“Adelie企鹅样本的嘴长度分布”
# 把图表的X轴标签设置为“嘴长度（单位：mm）”
# 把图表的Y轴标签设置为“样本数量”
plt.rcParams['font.sans-serif'] = ['SimHei']  # 或 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
sns.histplot(df1,x="bill_length_mm")
plt.title("Adelie企鹅样本的嘴长度分布")
plt.xlabel("嘴长度（单位：mm）")
plt.ylabel("样本数量")
plt.show()
```
