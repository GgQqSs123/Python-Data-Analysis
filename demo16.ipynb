```python
# 清洗数据
# 丢失数据不影响不处理 实在不行中位数平均数众数填充进去
# 清理步骤1人工填入缺失值2不处理缺失值3把有缺失值的行删除4用例如平均数等填充代替缺失值
# 针对重复数据找到删除
# 读取示例数据
import pandas as pd
df1 = pd.read_csv("example1.csv", index_col=0)
```

```python
df1
```

```python
# 重命名列和索引
# rename(index={})
# rename(columns={})
# 针对索引
df1=df1.rename(index={"2_":"2","_5":"5","6*":"6"})
```

```python
df1=df1.rename(columns={"客户_姓名":"客户姓名"})
```

```python
df1
```

```python
# 这是一种修改方法
# 另一种修改方法在 方法里面加上 inplace=True
df1.rename(index={"2_":"2","_5":"5","6":"7"},inplace =True)
```

```python
df1
```

```python
df1.rename(columns={"客户 性别":"性别"},inplace=True)
```

```python
df1
```

```python
#rename传回的是方法或值
df1.rename(columns=str.upper)
```

```python
# 把某列设置为索引
df1.set_index("性别")
```

```python
# 把索引重新设置为初始默认的位置索引
df1.reset_index()
```

```python
# 对索引和列名重新排序
# axis=0 沿着索引纵向操作 都不会改变原始dataframe所以可以家参数 inplace=True
df1.sort_index()
```
